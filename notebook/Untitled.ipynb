```python
barrier_keywords = (
    "barrier|reason for not having an account|no account|"
    "lack of money|too expensive|too far|family member has|"
    "lack of documentation|lack of trust|religious|no need|not useful|"
    "do not trust financial institutions"
)

# Filter only rows where the 'Indicator' contains barrier-related keywords
df_barriers = df[df["Indicator"].str.contains(barrier_keywords, case=False, regex=True)].copy()

# Optional: preview the result
print(df_barriers["Indicator"].unique())  # To see what barrier indicators are included
print(df_barriers.head())
```

```python
# Further filter for Rwanda only
df_rwanda_barriers = df_barriers[df_barriers["Country name"] == "Rwanda"]
```
